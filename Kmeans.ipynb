#Kmeans algorithm

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from random import randrange
```

```python
hw4data = loadmat('hw4_data.mat')
```

```python
hw4data
```

```python
sample_space = hw4data['X']
```

```python

k = 2
c = []
for i in range(k):
  c.append(sample_space[randrange(sample_space.shape[0])])#initialize 3 centroids randomly
jint = [0,0] #initialize distortion function to zero
while True:
  c1=[]       #intialize centroid labels array for samples 
  c2=[]       
  
  for i in range(sample_space.shape[0]):
    min=np.inf
    for j in range(len(c)):
      dist = np.linalg.norm(sample_space[i]-c[j]) #check distance of sample from given centroid
      if dist<min:
        min = dist
        k=j
    if k ==0:
      c1.append(sample_space[i])   #append samples to appropriate labels 
    else:
      c2.append(sample_space[i])
    
  c[0]= (np.sum(c1,axis=0)/len(c1)) #calculate new centroid by taking average
  c[1]= (np.sum(c2,axis=0)/len(c2))
  
  j = 0             #calculate new distortion function
  for i in c1:
    j += (i-c[0])**2
  for i in c2:
    j += (i-c[1])**2
  

  if j[0] == jint[0] and j[1]==jint[1]:# if distortion function value is same as previous then stop training.
    break
  else:
    jint = j

x=[]
y=[]
for i in c1:
  x.append(i[0])
  y.append(i[1])
plt.scatter(x, y, c ="blue")
x=[]
y=[]
for i in c2:
  x.append(i[0])
  y.append(i[1])
plt.scatter(x, y, c ="red")

x=[c[0][0],c[1][0]]
y=[c[0][1],c[1][1]]
plt.scatter(x, y, c ="black")
plt.show()
print("Number of blue points are {}".format(len(c1)))
print("The centroid of the blue points are {} and {}".format(c[0][0],c[0][1]))
print("Number of red points are {}".format(len(c2)))
print("The centroid of the red points are {} and {}".format(c[1][0],c[1][1]))
```

#Kmeans with sklearn library

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(sample_space)
x_red=[]
y_red=[]
x_blue=[]
y_blue=[]
for i in range(len(kmeans.labels_)):
  if kmeans.labels_[i] ==1:
    x_red.append(sample_space[i][0])
    y_red.append(sample_space[i][1])
  else:
    x_blue.append(sample_space[i][0])
    y_blue.append(sample_space[i][1])
plt.scatter(x_red, y_red, c ="red")
plt.scatter(x_blue, y_blue, c ="blue")

plt.show()
  
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(sample_space)
x_red=[]
y_red=[]
x_blue=[]
y_blue=[]
x_green=[]
y_green=[]

for i in range(len(kmeans.labels_)):
  if kmeans.labels_[i] ==0:
    x_red.append(sample_space[i][0])
    y_red.append(sample_space[i][1])
  elif kmeans.labels_[i] ==1:
    x_blue.append(sample_space[i][0])
    y_blue.append(sample_space[i][1])
  else: 
    x_green.append(sample_space[i][0])
    y_green.append(sample_space[i][1])

plt.scatter(x_red, y_red, c ="red")
plt.scatter(x_blue, y_blue, c ="blue")
plt.scatter(x_green, y_green, c ="green")
plt.show()
```
